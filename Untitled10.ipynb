```python
class adl:
  def __init__(self,s):
    self.s = s
  def dsf(self):
    r =[]
    pairs = {')': '(', '}': '{', ']': '['}
    for i in r:
      if i in pairs.values():
        r.append(i)
      else:
        if len(r) == 0 or r[-1] != pairs[i]:
          return False
        r.pop()
    return len(r) == 0
obj=adl("{)")
print(obj.dsf())
```
